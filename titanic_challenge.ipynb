```python
train_data = pd.read_csv("/kaggle/input/titanic-challenge/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic-challenge/test.csv")
test_data.head()
```

```python
import seaborn as sns
sns.heatmap(train_data[["Survived","Pclass", "Age", "SibSp", "Parch", "Fare"]].corr())
```

```python
Pclass = train_data.loc[train_data.Pclass == 1]["Survived"]
Pclass_rate = Pclass.value_counts()[1] / len(Pclass)

print("If you were in first class, the probability you survived was: ", Pclass_rate)
```

```python
Age = train_data.loc[train_data.Age < 10]["Survived"]
Age_rate = sum(Age) / len(Age)

print("If you were less than 10 years old, the probability you survived was ", Age_rate)
```

```python
Sex = train_data.loc[train_data.Sex == "female"]["Survived"]
Female_rate = sum(Sex) / len(Sex)

print("If you were a female, the probability that you survived was : ", Female_rate)
```

```python
features = ["Sex", "Age", "Pclass", "SibSp", "Parch", "Fare"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy="mean")

X = imp.fit(X).transform(X)
X_test = imp.fit(X_test).transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
y = train_data ["Survived"]
model = RandomForestClassifier()
model.fit(X,y)
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({"PassengerId": test_data.PassengerId, "Survived": predictions})
output.to_csv("submission.csv", index=False)

for file in os.listdir("/kaggle/working"):
    print(file)
```
